```python
import pandas as pd
```

```python
df24 = pd.read_csv('/pampa/cloutier/df24_consec.csv')
```

### Create csv file for each season

```python
seasons = {
    'SON': [9, 10, 11],
    'DJF': [12, 1, 2],
    'MAM': [3, 4, 5],
    'JJA': [6, 7, 8]
}

# Extract the month from the datetime column and create new column
df24['month'] = df24['datetime'].astype(str).str[4:6].astype(int)  

# dictionnary that will store the season in which a storm 
# has the highest grid point count
storm_seasons = {}
for storm_id, storm_data in df24.groupby('storm'):
    
    # dictionnary that will store month count for each season
    season_counts = {season: 0 for season in seasons}
    
    for season, months in seasons.items():
        # count the total of grid point in each season
        season_counts[season] = len(storm_data[storm_data['month'].isin(months)])
        
    # The storm will be added in the season in which the count was the highest
    max_season = max(season_counts, key=season_counts.get)
    storm_seasons[storm_id] = max_season

# Separate the dataframe by season and save to CSV files
for season, months in seasons.items():
    season_df = df24[df24['month'].isin(months)]
    season_df = season_df[season_df['storm'].map(storm_seasons) == season]
    #season_csv = f'etc_24_{season}.csv'
    #season_df.to_csv(season_csv, index=False)
```
