```python
# import filtered ETCs
fetc = pd.read_csv("/pampa/cloutier/storm_tracks/filtered/etc24_consec_season.csv")
```

```python
# get the average over the duration when the center was within the CRCM6 region
fetc_true = fetc.loc[fetc.HU == True].copy()
fetc_avg = fetc_true.groupby(['storm', 'season']).agg({'85WS_av08': 'mean', '10WS_av08': 'mean'}).reset_index()

mean_10 = fetc_avg['10WS_av08'].mean()
mean_85 = fetc_avg['85WS_av08'].mean()
std_10 = fetc_avg['10WS_av08'].std()
std_85 = fetc_avg['85WS_av08'].std()

# normalize the values of fetc_avg
fetc_avg['n_85WS_av08'] = (fetc_avg['85WS_av08'] - mean_85) / std_85
fetc_avg['n_10WS_av08'] = (fetc_avg['10WS_av08'] - mean_10) / std_10
```

```python
# separate data by season
seasons = ['JJA', 'SON', 'DJF', 'MAM']
columns = ['n_85WS_av08', 'n_10WS_av08']

# Loop through each season and column suffix to extract the data and create arrays
for season in seasons:
    for suffix in columns:
        column_name = f"{suffix}_{season.lower()}"
        array_name = f"{suffix}_{season.lower()}"
        exec(f"{array_name} = fetc_avg.loc[fetc_avg['season'] == '{season}']['{suffix}']")
```

```python
# count the number of occurences per season and per quadrant

quadrant = ['1', '2', '3', '4']
seasons = ['JJA', 'SON', 'DJF', 'MAM']

for q in quadrant : 
    for s in seasons : 
        var = f"{s.lower()}_{q}"
        if q == '1' : 
            exec(f"{var} = len(fetc_avg.loc[(fetc_avg['season'] == '{s}') & (fetc_avg['n_85WS_av08'] > 1) & (fetc_avg['n_10WS_av08'] > 1)])")
            
        if q == '2' : 
            exec(f"{var} = len(fetc_avg.loc[(fetc_avg['season'] == '{s}') & (fetc_avg['n_85WS_av08'] < 1) & (fetc_avg['n_10WS_av08'] > 1)])")
        
        if q == '3' : 
            exec(f"{var} = len(fetc_avg.loc[(fetc_avg['season'] == '{s}') & (fetc_avg['n_85WS_av08'] <= 1) & (fetc_avg['n_10WS_av08'] <= 1)])")
            
        if q == '4' : 
            exec(f"{var} = len(fetc_avg.loc[(fetc_avg['season'] == '{s}') & (fetc_avg['n_85WS_av08'] > 1) & (fetc_avg['n_10WS_av08'] < 1)])")
```
