```python
# Calculer la distance entre les points de grille consécutifs pour chaque tempête et regarder si
# certaines tempêtes se sont déplacées de plus de un point de grille (0.25°) en un pas de temps. 
```

```python
cat = pd.read_csv('/home/data/ReAnalysis/ERA5/Storm_analysis/NAECv1/NAEC_1979_2020_v1.csv')
```

```python
grouped_df = cat.groupby('storm')

result_df = pd.DataFrame()

for storm_name, group in grouped_df:
    lat_diff = group[['latitude']].diff().abs()
    lon_diff = group[['longitude']].diff().abs()
    storm_max_lat_diff = lat_diff['latitude'].max()
    storm_max_lon_diff = lon_diff['longitude'].max()
    print(storm_max_lon_diff)
    print(storm_max_lat_diff)

#     # Update the original group with the differences
#     group[['latitude', 'longitude']] = lat_lon_diff

#     # Append the updated group to the result DataFrame
#     result_df = pd.concat([result_df, group])
```
