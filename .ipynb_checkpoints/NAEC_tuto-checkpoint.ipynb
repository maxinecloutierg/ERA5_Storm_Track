### Create map for january 1rst 2020 tracks

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import numpy as np

df2 = df.loc[(df.datetime//100) == 20000101]
#print(df2)
```

```python
# Les 4 coordonnes du domaine de la Figure 1 : 
# upper left : 62.922848, -131.799332
# lower left : 28 -109
# upper right : 52 -16
# lower right : 16.204049, -38.233449
# milieu de la map : 55 -87

m = Basemap(llcrnrlon=-109, llcrnrlat=28, urcrnrlon=-15, urcrnrlat=45, resolution='i', lat_0 = 60, lon_0 = -90,  
            area_thresh=1500.,projection='lcc')

fig = plt.figure(figsize=(12,9))

# drawing the coastline
m.drawcoastlines()
m.drawcountries(color='gray')
m.drawstates(color='gray')

# creating variable for latitude and longitude to list
lat = df2['latitude'].tolist()
lon = df2['longitude'].tolist()

m.scatter(lon, lat, latlon = True, c = 'red', marker = 'o', s=10, alpha=1)

plt.show()
```
